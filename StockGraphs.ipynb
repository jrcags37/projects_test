```python
import json
import urllib.request
import urllib.parse
import ipywidgets as widgets
import re
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
plt.style.use('seaborn-whitegrid')
plt.rcParams.update({'font.size': 22})
figure(figsize=(20, 10), dpi=240)
figure(2, figsize=(20, 10), dpi=240)
figure(3, figsize=(20, 10), dpi=240)
figure(4, figsize=(20, 10), dpi=240)



site = 'https://finnhub.io/api/v1/stock/metric'
stock = input('Enter a ticker symbol: ')
query_sym = '?'
paramsA = urllib.parse.urlencode({'token': 'c8pn242ad3icps1js550', 'symbol': stock})
response_object = urllib.request.urlopen(site + query_sym + paramsA)
response_string = response_object.read()
response_as_dictionary = json.loads(response_string)

#EARNINGS PER SHARE
epsinfo = response_as_dictionary['series']['annual']['eps']

EPSx = []
EPSy = []

for item in epsinfo:
    EPSx.append(item['period'][2:4]) 
    EPSy.append(item['v'])

EPSx.reverse()
EPSy.reverse()

plot1 = plt.figure(1)
plt.plot(EPSx, EPSy)
plt.title(f'{stock} Earnings Per Share')
plt.xlabel('Period')
plt.ylabel('EPS')

#OPERATING MARGIN
opmargin = response_as_dictionary['series']['annual']['operatingMargin']
OpMx = []
OpMy = []

for item in opmargin:
    OpMx.append(item['period'][2:4]) 
    OpMy.append(item['v'])

OpMx.reverse()
OpMy.reverse()

plot2 = plt.figure(2)
plt.plot(OpMx, OpMy)
plt.title(f'{stock} Operating Margin')
plt.xlabel('Period')
plt.ylabel('Margin')

#SG&A/Sales
sga = response_as_dictionary['series']['annual']['sgaToSale']
sgax = []
sgay = []

for item in sga:
    sgax.append(item['period'][2:4]) 
    sgay.append(item['v'])

sgax.reverse()
sgay.reverse()

plot4 = plt.figure(4)
plt.plot(sgax, sgay)
plt.title(f'{stock} SG&A to Sales')
plt.xlabel('Period')
plt.ylabel('SG&A/Sales')

#Debt to Equity
de = response_as_dictionary['series']['annual']['totalDebtToEquity']
dex = []
dey = []

for item in de:
    dex.append(item['period'][2:4]) 
    dey.append(item['v'])

dex.reverse()
dey.reverse()

plot3 = plt.figure(3)
plt.plot(dex, dey)
plt.title(f'{stock} Total Debt/Equity')
plt.xlabel('Period')
plt.ylabel('D/E')

plt.show()
```
